# Vectors & Linear Combinations

In this notebook, will learn how to plot the two dimentional vectors using `quiver` plot.

**What is a vector?**

Any such entity involving both magnitude and direction is called a “vector.” For a physicist, a vector is represented by an arrow whose length denotes the magnitude of the vector and whose direction represents the direction of the vector. For a computer scientist, a vector is an array of numbers. For a mathematician, vector is an element of a vector space. Before going into the general abstract definition of a vector space, we will first of all take a look at the vectors of $\mathbb{R}^2$ and the basic operations of vectors.

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
plt.quiver(0,0,1,1,
           scale=1,
           angles='xy',
           scale_units='xy')
plt.ylim(0,1.2)
plt.xlim(0,1.2)
```

```python
# helper function to plot the vectors
def plot_vectors(vectors, title=None):
    import numpy as np
    from matplotlib import pyplot as plt
    # Initiate the plot
    fig, ax = plt.subplots(figsize = (5,5))
    
    # Draw each vector as quiver plot
    for x_cor, y_cor in vectors:
        ax.quiver(0,0,x_cor,y_cor, #note that 0,0 gives the tail of the vector
                  scale=1,angles='xy',
                  scale_units='xy',
                  color='steelblue'
                 )
    # Calculate the limit
    limit = np.max(np.abs(vectors))*1.25
    ax.set_xlim([-limit, limit])
    ax.set_ylim([-limit, limit])
    ax.set_aspect('equal')
    
    # Make the grid visible
    ax.grid(True, linewidth = 0.5, alpha= 0.85)
    
    # show x-y axis in the center, and hide the frames
    ax.spines['left'].set_position('center')
    ax.spines['bottom'].set_position('center')
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    if title != None:
        plt.title(title)
    plt.show()
```

Now, let us use the function `plot_vectors` to plot our vectors to check everything works as expected.

```python
V = [(-1,2), (1,3), (3,4)]
plot_vectors(vectors = V, title="Plotting vectors using quiver plot")
```

## Plotting the span of two vectors

We know that the span of two vectors $a$ and $b$ is defined as the set of all linear combination of $a$ and $b$. That is,
$$ span(\{a,b\} = \{ \alpha a + \beta b : \alpha, \beta\in \mathbb R\} $$

Let us randomly select some scalars and plot the linear combination and try to visualize the span approximately.

```python
import random
a = np.array((0,1)); b = np.array((1,0))
noLCs = 100 # no of linear combinations to be generated
lc = []
for _ in range(noLCs):
    alpha = random.randint(-20,20)
    beta = random.randint(-20,20)
    lc.append(alpha * a + beta * b)
plot_vectors(lc, title=f"Span of a, b ({noLCs} linear combinations)")
```
